```python
import pandas as pd
import json
import ast
from pandas.io.json import json_normalize
```

```python
df = pd.read_csv("flats_df.csv")
```

```python
df.dropna(axis=1, how='all', inplace=True)
```

```python
df.drop(['added', 'addedTimestamp', 'bedsCount', 'categoriesIds', 'cianId', 'cianUserId',
         'descriptionMinhash', 'descriptionWordsHighlighted', 'forDay', 'gaLabel', 'gaObjectType',
         'jkUrl', 'modelVersion', 'objectGuid', 'platform', 'publishedUserId', 'specialty', 'userId',
         'videos', 'photos', 'windowsViewType', 'withoutClientFee', 'adfoxParams', 'humanizedTimedelta',
         'id', 'newbuildingVasPromotion', 'notes'], inplace=True, axis=1)
```

# ==============================================================
# <><><><><><><><><><>Extracting data from neted dicts<><><><><><><><>
# ==============================================================

```python
def extract_key(json_string, key):
    return eval(json_string)[key]

def flatten_json(d):
    return ast.literal_eval(d)

def flatten_list_of_dicts(ld):
    return dict([(list(d.values())[1], list(d.values())[0]) for d in ast.literal_eval(ld)])

def convert_to_list(obj):
    if type(obj) == str:
        return eval(obj)
    else:
        return obj

def convert_to_str(obj):
    if type(obj) == str:
        return obj
    else:
        return str(obj)
```

```python
df['allRoomsArea'].fillna('0', inplace=True)
```

```python
df['balconiesCount'] = df['balconiesCount'].fillna(0).astype(int)
```

```python
df['flatPrice'] = df['bargainTerms'].apply(extract_key, key='price')
df['flatPriceCurrency'] = df['bargainTerms'].apply(extract_key, key='currency')
df['flatPriceDeposit'] = df['bargainTerms'].apply(extract_key, key='deposit')

df.drop('bargainTerms', inplace=True, axis=1)
```

```python
df['buildYear'] = df['building'].apply(extract_key, key='buildYear')
df['buildMaterialType'] = df['building'].apply(extract_key, key='materialType')
df['buildfloorsCount'] = df['building'].apply(extract_key, key='floorsCount')
df['buildPassengerLiftsCount'] = df['building'].apply(extract_key, key='passengerLiftsCount')
df['buildCargoLiftsCount'] =  df['building'].apply(extract_key, key='cargoLiftsCount')

df.drop('building', inplace=True, axis=1)
```

```python
# need to resolve 'geo'
```

```python
json_normalize(df['geo'].apply(flatten_json).tolist()).add_prefix('geo.')
```

```python
df['newbuilding'].fillna(False, inplace=True)
df.loc[df['newbuilding'] != False, 'newbuilding'] = True

df.drop('newbuilding', inplace=True, axis=1)
```

```python
df['user'].fillna("{}", inplace=True)
```

```python
agency_data = json_normalize(df['user'].apply(flatten_json).tolist()).add_prefix('user.')
agency_data['user.phoneNumbers'].fillna("[{'number': '0', 'countryCode': '+7'}]", inplace=True)
```

```python
agency_data.head()
```

```python
df['userAccountType'] = agency_data['user.accountType']
```

```python
df['userAgencyName'] = agency_data['user.agencyName']
```

```python
df['userAgentAvailabilityUserId'] = agency_data['user.agentAvailability.userId']
```

```python
df['userPhoneNumber'] = agency_data['user.phoneNumbers'].apply(convert_to_list).apply(lambda x: x[0]['countryCode'] + x[0]['number'])
```

```python
df['creationDate'] = pd.to_datetime(df['creationDate'])
```

```python
df.drop(['phones', 'user'], inplace=True, axis=1)
```

```python
df.to_csv('flat_data_v2_4122019.csv', index=False, encoding='utf-8')
```
